```typescript
import { PromptTemplate } from "@langchain/core/prompts";

// 自动从模板中推断变量
const autoInferTemplate = PromptTemplate.fromTemplate("good {timeOfDay}, {name}!");
console.log(autoInferTemplate.inputVariables); // Output: ['timeOfDay', 'name']

const formattedAutoInferTemplate = await autoInferTemplate.format({ timeOfDay: "evening", name: "Jaycee" });
console.log(formattedAutoInferTemplate); // Output: good evening, Jaycee!
```
