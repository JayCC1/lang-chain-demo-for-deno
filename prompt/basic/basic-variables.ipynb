```typescript
import { PromptTemplate } from "@langchain/core/prompts";

// 如果 prompt 需要 {}，可以这么转义{{}}
const personalizedGreetingPrompt = new PromptTemplate({
    inputVariables: ["timeOfDay", "name"],
    template: "Good {timeOfDay}, {name} {{test}}!",
});

const formattedPersonalizedGreeting = await personalizedGreetingPrompt.format({ timeOfDay: "morning", name: "Jaycee" });
console.log(formattedPersonalizedGreeting);  // Output: Good morning, Jaycee {test}!
```
