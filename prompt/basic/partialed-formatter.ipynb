```typescript
import { PromptTemplate } from "@langchain/core/prompts";

const initialPrompt = new PromptTemplate({
    template: "这是一个{type}，它是{item}",
    inputVariables: ["type", "item"],
});

// 不需要一次性把所有变量都输入进去
// 在工程中，我们可能先获得某个参数，之后才能获得另一个参数
// 这时可以使用 partial 方法，先填充部分变量
const partialedPrompt = await initialPrompt.partial({ type: "工具" });

const formattedPrompt = await partialedPrompt.format({ item: "锤子" });
console.log(formattedPrompt); // Output: 这是一个工具，它是锤子

// 之后还可以继续使用同一个 partialedPrompt，填充不同的 item
const formattedPrompt2 = await partialedPrompt.format({ item: "螺丝刀" });
console.log(formattedPrompt2); // Output: 这是一个工具，它是螺丝刀
```
